```python
import numpy as np
from wgpu.gui.auto import WgpuCanvas
import pygfx as gfx
from gridplot import GridPlot, Image
```

```python
canvas = WgpuCanvas()
renderer = gfx.renderers.WgpuRenderer(canvas)

images = [(np.random.rand(512, 512) * 255).astype(np.float32) for i in range(4)]

grid_plot = GridPlot(
    canvas=canvas,
    renderer=renderer,
    grid_shape=(2, 2),
    cameras=np.array([['o', 'o'], ['o', 'o']]),
    controllers=np.array([[0, 1], [2, 3]])
)

for subplot, img in zip(grid_plot, images):
    subplot.add_graphic(Image(data=img, vmin=0, vmax=255))
    
canvas.request_draw(grid_plot.animate)
canvas
```
